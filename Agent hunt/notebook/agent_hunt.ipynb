### Normalizing the data

#### Brokerage df

```python
brokerage_df.columns
```

```python
brokerage_df['name'] = brokerage_df['name'].str.lower()
brokerage_df['email'] = brokerage_df['email'].str.lower()
brokerage_df['street'] = brokerage_df['street'].str.lower()
brokerage_df['city'] = brokerage_df['city'].str.lower()
brokerage_df
```

```python
brokerage_df.drop_duplicates(subset="name",keep=False, inplace=True)
brokerage_df
```

```python
brokerage_df.drop_duplicates(subset="phone_numbers",keep=False, inplace=True)
brokerage_df
```

```python
brokerage_df.drop_duplicates(subset=["street","city","county","state","zipcode"],keep=False, inplace=True)
brokerage_df = brokerage_df.reset_index(drop=True)
brokerage_df
```

### Agent_info

```python
agent_info_df.columns
```

```python
agent_info_df['first_name'] = agent_info_df['first_name'].str.lower()
agent_info_df['email'] = agent_info_df['email'].str.lower()
agent_info_df['street'] = agent_info_df['street'].str.lower()
agent_info_df['city'] = agent_info_df['city'].str.lower()
agent_info_df['state'] = agent_info_df['state'].str.lower()
```

```python
agent_info_df.drop_duplicates(subset="state_license",keep=False, inplace=True)
agent_info_df.drop_duplicates(subset="first_name",keep=False, inplace=True)
agent_info_df.drop_duplicates(subset=["street","city","state","zipcode"],keep=False, inplace=True)
agent_info_df= agent_info_df.reset_index(drop=True)
agent_info_df 
```
