```python
import pandas as pd
agent_info_df = pd.read_sql('SELECT * FROM agent_info',engine)
agent_info_df.head()
```

```python
brokerage_df = pd.read_sql('SELECT * FROM brokerage',engine)
brokerage_df.head()
```

```python
home_info_df = pd.read_sql('SELECT * FROM home_info',engine)
home_info_df.head()
```

```python
agent_listing_df = pd.read_sql('SELECT * FROM agent_listing',engine)
agent_listing_df.head()
```

### Normalizing the data

#### Brokerage df

```python
brokerage_df.columns
```

```python
brokerage_df['name'] = brokerage_df['name'].str.lower()
brokerage_df['email'] = brokerage_df['email'].str.lower()
brokerage_df['street'] = brokerage_df['street'].str.lower()
brokerage_df['city'] = brokerage_df['city'].str.lower()
brokerage_df
```

```python
brokerage_df.drop_duplicates(subset="name",keep=False, inplace=True)
brokerage_df
```

```python
!pip install fuzzywuzzy
```

```python
brokerage_df.drop_duplicates(subset="phone_numbers",keep=False, inplace=True)
brokerage_df
```

```python
brokerage_df.drop_duplicates(subset=["street","city","county","state","zipcode"],keep=False, inplace=True)
brokerage_df = brokerage_df.reset_index(drop=True)
brokerage_df
```

### Agent_info

```python
agent_info_df.columns
```

```python
agent_info_df['first_name'] = agent_info_df['first_name'].str.lower()
agent_info_df['email'] = agent_info_df['email'].str.lower()
agent_info_df['street'] = agent_info_df['street'].str.lower()
agent_info_df['city'] = agent_info_df['city'].str.lower()
agent_info_df['state'] = agent_info_df['state'].str.lower()
```

```python
agent_info_df.drop_duplicates(subset="state_license",keep=False, inplace=True)
agent_info_df.drop_duplicates(subset="first_name",keep=False, inplace=True)
agent_info_df.drop_duplicates(subset=["street","city","state","zipcode"],keep=False, inplace=True)
agent_info_df= agent_info_df.reset_index(drop=True)
agent_info_df['brokerage_id'] = pd.to_numeric(agent_info_df['brokerage_id'], errors='coerce')
agent_info_df['brokerage_id'].fillna(0, inplace=True)
agent_info_df['brokerage_id'] = agent_info_df['brokerage_id'].astype(int)
agent_info_df 
```

```python
mask = agent_info_df['brokerage_id'].isin(brokerage_df['id'])
agent_info_df = agent_info_df[mask]
agent_info_df
```

### Agent_listing

```python
mask = agent_listing_df['agent_id'].isin(agent_info_df['id'])
agent_listing_df = agent_listing_df[mask]
agent_listing_df
```

### home_info

```python
home_info_df
```

```python
home_info_df.fillna(home_info_df.mean(),inplace=True)
home_info_df = home_info_df.applymap(lambda x: int(x) if isinstance(x, float) else x)
home_info_df
```

```python
engine1 = create_engine('postgresql://postgres:root@localhost:5432/postgres')
```

```python
home_info_df.to_sql('home_info', engine1, if_exists='replace')
agent_listing_df.to_sql('agent_listing', engine1, if_exists='replace')
agent_info_df.to_sql('agent_info', engine1, if_exists='replace')
brokerage_df.to_sql('brokerage', engine, if_exists='replace')
```
